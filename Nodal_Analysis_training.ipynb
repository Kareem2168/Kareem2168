```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
```

```python
q_ipr=[0, 1999, 3094, 3902, 4512, 4963, 5275, 5458, 5519]
pwf_ipr=[4000,  3500, 3000, 2500, 2000, 1500, 1000, 500, 14.65 ]
```

```python
IPR=pd.DataFrame({'Q':q_ipr, 'Pwf':pwf_ipr})
IPR
```

```python
plt.plot(IPR['Q'], IPR['Pwf'])
```

```python
#VFP
q_vfp=np.arange(1000, 5500, 500)
q_vfp
```

```python
p_190=[1334,1400,1498,1503,1590,1609,1659,1798,1859]
p_2875=[1286, 1302,1349,1356,1389,1398,1420,1438,1444]
```

```python
plt.plot(IPR['Q'], IPR['Pwf'], label='IPR')
plt.plot(q_vfp,p_190, label='1.9inch' )
plt.plot(q_vfp, p_2875, label='2.875inch')
plt.legend()
```

using plotly

```python
import plotly.graph_objs as go 



```

```python
data=go.Scatter(x=IPR['Q'],y= IPR['Pwf'], name='IPR')
data1=go.Scatter(x=q_vfp,y=p_190, name='1.9 inch')
layout=go.Layout(title='nodal analysis')
figure=go.Figure(data=(data,data1), layout=layout)
figure
```

lets see another nodal analysis example 

```python
e=np.linspace(0,1.2,50)
for i in e:
 if i>=1:

   data=go.Scatter(x=IPR['Q'],y= IPR['Pwf'], name='IPR')
   data1=go.Scatter(x=q_vfp,y=p_190, name='1.9 inch')
   data2=go.Scatter(x=e, name='EV')
   layout=go.Layout(title='nodal analysis 1')
   figure1=go.Figure(data=(data,data1, data2), layout=layout)
   figure1.show()

 else:
    data=go.Scatter(x=IPR['Q'],y= IPR['Pwf'], name='IPR')
    data1=go.Scatter(x=q_vfp,y=p_190, name='1.9 inch')
    
    layout=go.Layout(title='nodal analysis')
    figure2=go.Figure(data=(data,data1), layout=layout)
    figure2.show()
```
